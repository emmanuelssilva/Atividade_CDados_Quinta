Contexto dos Dados.

Esses dados foram extraidos após o episódio "The Rams Are Dead To Me"(Os Rams morreram para mim), o time de futebol americano Rams se mudou de St. Louis para Los Angeles, causando uma enorme decepção para os torcedores da cidade. 
Tendo esse problema, uma pesquisa foi realizada com 3.352 pessoas levantando os pontos que influenciaria as pessoas da cidade a mudar de time.

> Colunas:

* FRL - Relações com os torcedores - cortesia de jogadores, treinadores e front office para os torcedores, e quão bem um time usa a tecnologia para alcançá-los.

* OWN - Propriedade - Honestidade com a comunidade, lealdade dos principais jogadores à comunidade.

* PLA - Jogadores - esforço em campo, simpatia fora dele.

* FUT - Vitórias futuras - vitórias projetadas nas próximas 5 temporadas.

* BWG - Fator do movimento - os próximos 5 anos da equipe provavelmente serão melhores do que os 5 anteriores?

* TRD - Tradição - Super Bowls Vencidos / Títulos de Conferência e divisão / Vitórias em toda a história da equipe.

* BNG - Cuidado com os investimentos - garantia de que o dinheiro investido dos torcedores serão revertidos para o clube.

* BEH - Comportamento - Suspensão de jogadores da equipe desde 2007, Jogadores com o comportamento inadequado fora dos campos. 

* NYP - Proximidade com a cidade de Nova York.

* SLP - Proximidade de St. Louis.

* AFF - Acessibilidade - Preço das passagens, estacionamento e concessões.

* SMK - Clube Pequeno - Tamanho do clube em termos de torcida, onde menor é melhor.

* STX - Experiência no estádio - Qualidade do local, ambiente favorável aos fãs, frequência das promoções do dia do jogo.

* CCH - Coaching(Técnico) - Força da liderança em campo.
* UNI - Uniforme - Estilo de design uniforme, de acordo com Paul Lukas da Uni Watch.

* BMK - Grande clube - Tamanho do clube em termos de torcida, onde maior é melhor.







Os dados foram levantados no ano de 2016.

# **0 - Importação das Bibliotecas Pandas:**

```python
# Importação das bibiotecas.
import pandas as pd
import numpy as np
```

# **Importação dos Dados:**

```python
# Importando o arquivo csv.
nfl_df = pd.read_csv("team-picking-categories.csv")
```

# **1 - Leitura de dados com pandas:**

```python
# Exibindo resultado.
display(nfl_df)
```

# **2 - Manipulação e limpeza de dados com pandas:**

```python
# Excluindo colunas que serão utilizadas.
nfl_df = nfl_df.drop("BWG", axis=1)

# Deletando linhas que possuem pelo menos 1 valor vazio.
nfl_df = nfl_df.dropna()

# Exibindo resultado da manipulação e limpeza.
display(nfl_df)
```

# **3 - Descritivo e exploração de dados com pandas (groupbys):**

```python
# Listando as pessoas que moram na região central de St. Louis.
time_stlouis = nfl_df[["TEAM", "SLP"]].groupby("TEAM").sum()
display(time_stlouis)
```

```python
# Listando as pessoas que torceriam por conta da tradição do time.
time_tradicao = nfl_df[["TEAM", "TRD"]].groupby("TEAM").sum()
display(time_tradicao)
```

# **5 - Quebrar os dados em conjunto de treino e conjunto de teste:**

```python
# Adicionando a função do scikit-learn.
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
```

```python
# Excluindo a coluna country para a realização dos teste.
nfl_df = nfl_df.drop("TEAM", axis=1)
```

```python
# Criando uma mascara com as posicões.
msk = np.random.rand(len(nfl_df)) < 0.3
# Definindo as posições.
X = nfl_df[msk]
Y = nfl_df[~msk]
```

```python
# Verificando o tamanho do corpo do X.
X.shape
```

```python
# Verificando o tamanho do corpo do Y.
Y.shape
```

```python
# Convertendo argumentos strings em tipo númerico.
X = X.apply(pd.to_numeric, errors='coerce')
Y = Y.apply(pd.to_numeric, errors='coerce')
```

```python
# Preenchendo os valores vazios no X e no Y
X.fillna(0, inplace=True)
Y.fillna(0, inplace=True)
```

# **6 - Treinar um estimador do scikit-learn de aprendizagem supervisionada (regressor ou classificador) no conjunto de treino:**

```python
# Retirando colunas com variaveis categoricas
X = nfl_df.drop("TRD", axis=1)
Y = nfl_df["TRD"]
```

```python
# Separando os dados de para teste e treino.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
# Criando e Treinando os modelos.
regr = LinearRegression()
regr.fit(X_train, Y_train)
```

# **7 - Verificar o desempenho do estimador do scikit-learn no conjunto de teste:**

```python
# Imprimindo resultados.
r2_train = regr.score(X_train, Y_train)
r2_test = regr.score(X_test, Y_test)
print('R2 no set de treino: %.2f' % r2_train)
print('R2 no set de teste: %.2f' % r2_test)
```
